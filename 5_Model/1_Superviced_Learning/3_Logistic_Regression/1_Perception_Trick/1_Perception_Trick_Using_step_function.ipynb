```python
from sklearn.datasets import make_classification
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(
    n_samples=100,
    n_features=2,
    n_informative=1,
    n_classes=2,
    n_clusters_per_class=1,
    n_redundant=0,
    random_state=41,
    class_sep=30,
    hypercube=False
)
```
