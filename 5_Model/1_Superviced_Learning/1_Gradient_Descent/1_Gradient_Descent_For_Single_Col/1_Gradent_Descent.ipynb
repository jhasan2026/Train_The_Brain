```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
y
```

```python
plt.scatter(X,y)
```

# 1. Check Linear Reg alogo's regression line

```python
lr = LinearRegression()
lr.fit(X,y)
```

```python
intercept = lr.intercept_
coef =  (lr.coef_)[0]
intercept, coef
```

```python
y_pred = X*coef + intercept
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred, color='r')
```

# 2. Change In intercept

```python
intercept_1 = 100.0
intercept_1, coef
```

```python
y_pred_1 = coef*X + intercept_1
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred, color='r')
plt.plot(X,y_pred_1, color='g')
```

```python
print(X.shape)
print(X.ravel().shape)
```

```python
# iter = 1
dL_by_db = -2 * np.sum(y - coef*(X.ravel()) - intercept_1)
intercept_2 = intercept_1 - 0.1*dL_by_db
print(intercept_2)
y_pred_2 = coef*X + intercept_2
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred, color='r')
plt.plot(X,y_pred_1, color='g')
plt.plot(X,y_pred_2, color='purple')
```

```python
# iter = 2
dL_by_db = -2 * np.sum(y - coef*(X.ravel()) - intercept_2)
intercept_3 = intercept_2 - 0.1*dL_by_db
print(intercept_3)
y_pred_3 = coef*X + intercept_3
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred, color='r')
plt.plot(X,y_pred_1, color='g')
plt.plot(X,y_pred_2, color='purple')
plt.plot(X,y_pred_3, color='yellow')
```

```python
# iter-3 change in coef
n = X.shape[0]
Loss = y - (coef * X.ravel() + intercept_2)
dL_by_dw = -(2/n) * np.dot(X.ravel(), Loss)

coef_3 = intercept_2 - 0.1*dL_by_dw
print(coef_3)
y_pred_4 = coef_3*X + intercept_3
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred, color='r')
plt.plot(X,y_pred_1, color='g')
plt.plot(X,y_pred_2, color='purple')
plt.plot(X,y_pred_3, color='yellow')
plt.plot(X,y_pred_4, color='#18020C')
```

# Using Loop 

```python
import random
def random_color_generate():
    n = random.randint(0, 0xFFFFFF)
    return f'#{n:06x}'
```

```python
random_color_generate()
```

```python
epochs = 10
lr = 0.01
```

```python
def using_df(epochs, lr):
    coef = 78.35
    intercept = 100
    plt.scatter(X,y)
    for epoch in range(epochs):
        dL_by_db = -2 * np.sum(y - coef*(X.ravel()) - intercept)
        intercept = intercept_2 - 0.1*dL_by_db
        y_pred = coef*X + intercept
        plt.plot(X,y_pred, color='{}'.format(random_color_generate()), label=f'Epoch: {epoch}')
        
    plt.legend()
```

```python
using_df(10,0.01)
```
