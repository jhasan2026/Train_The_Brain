# Extract Numerical and String data

```python
df['numerical_part'] = pd.to_numeric(df['number'], errors='coerce',downcast='integer')
df
```

```python
df['string_part'] = np.where(df['numerical_part'].isnull(), df['number'], np.nan)
df
```

```python
df['Cabin_Str'] = df['Cabin'].str[0]
df['Cabin_Numeric'] = df['Cabin'].str.extract('(\d+)')
df
```
