```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../Data/titanic_Dataset.csv",usecols=['Name','Age','Pclass','SibSp','Parch','Survived'])
df
```

# 1. Feature Construction

```python
df['Family_size'] = df['SibSp'] + df['Parch']
df
```

```python
def set_family_type(size):
    if size <= 1:
        return 0
    elif size <= 4:
        return 1
    else:
        return 0
```

```python
df['Family_size'] = df['Family_size'].apply(set_family_type)
df
```

# 2. Split Feature

```python
df['name_title'] = df['Name'].str.split(", ", expand=True)[1].str.split(".",expand=True)[0]
df
```

```python
df['name_title'].value_counts()
```

```python
df['is_married'] = ((df['name_title'] == 'Mr') | (df['name_title']=='Mrs'))
```
