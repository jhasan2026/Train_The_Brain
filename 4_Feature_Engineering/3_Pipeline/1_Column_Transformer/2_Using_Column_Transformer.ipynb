```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../../Data/covid_toy.csv")
df
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df.iloc[:, -1], test_size=0.2, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
X_train
```

# 1. Fever -> Imputation

# 2. Age, Fever -> ScalingM

# 3. City. Gender -> OHE

# 4. cough -> Ordinal Encoding

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.impute import SimpleImputer
```

```python
transformer = ColumnTransformer(transformers=[
    ('trf1', SimpleImputer(), ['fever']),
    ('trf2', StandardScaler(), ['age','fever']),
    ('trf3', OneHotEncoder(drop='first',sparse_output=False, dtype=np.int32), ['city','gender']),
    ('trf4', OrdinalEncoder(categories=[['Mild','Strong']]), ['cough'])
],remainder='passthrough')
```

```python
X_train_trf = transformer.fit_transform(X_train)
X_test_trf = transformer.transform(X_test)
```
