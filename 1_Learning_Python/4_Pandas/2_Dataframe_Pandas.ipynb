# apply

```python
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)
points_df
```

```python
def euclidian_distance(row):
    p1 = row['1st point'] 
    p2 = row['2nd point'] 
    return ((p1[0] - p2[0])**2  + (p1[1] - p2[1])**2 )**0.5
```

```python
euclidian_distance(points_df.iloc[0])
```

```python
points_df.apply(euclidian_distance, axis=1)
```
