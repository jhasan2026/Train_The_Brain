# e) 3D Surface Plots
### (for plane)

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx , yy = np.meshgrid(x,y)
```

```python
xx
```

```python
# z = x^2 + y^2
z = xx**2 + yy**2
print(z.shape)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

sur_obj = ax.plot_surface(xx, yy , z, cmap='viridis') 
fig.colorbar(sur_obj)   # object for this colorbar
```

```python
# sinx
z = np.sin(xx)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

sur_obj = ax.plot_surface(xx, yy , z, cmap='viridis') 
fig.colorbar(sur_obj)   
```

```python
# z = sinx + cosy
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

sur_obj = ax.plot_surface(xx, yy , z, cmap='viridis') 
fig.colorbar(sur_obj)   
```

```python
# sinx + logy
z = np.sin(xx) + np.log(yy)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

sur_obj = ax.plot_surface(xx, yy , z, cmap='viridis') 
fig.colorbar(sur_obj)   
```

# g) Heatmap 
### (x -> cat) (y->cat) (color -> num)

```python
df = pd.read_csv('Data/IPL_Ball_by_Ball_2008_2022.csv',usecols=['overs','ballnumber','batsman_run'])
df
```

```python
temp_df = df[(df['ballnumber'].isin([1,2,3,4,5,6])) & (df['batsman_run'] == 6)]
temp_df
```

```python
# row-> index , cols =columns, color_code -> value, aggfunc
grid = temp_df.pivot_table(index='overs', columns='ballnumber', values='batsman_run', aggfunc='count')
```

```python
fig = plt.figure(figsize=(20,10))

plt.imshow(grid , cmap='viridis')
plt.colorbar()
```

# 4. Subplot

```python
batter
```

```python
# color and marker->point style
fig, ax = plt.subplots(figsize=(15,6))

ax.scatter(batter['avg'],batter['strike_rate'], color='red',marker='+')
ax.set_title("Avg vs Strike rate")
ax.set_xlabel("Avg")
ax.set_ylabel("Strike rate")
fig.show()
```

```python
# array (with x coordinate are same)
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(15,6))

ax[0].scatter(batter['avg'],batter['strike_rate'], color='red')
ax[0].set_title("Avg vs Strike rate")
ax[0].set_ylabel("Strike rate")

ax[1].scatter(batter['avg'],batter['runs'], color='blue')
ax[1].set_title("Avg vs Run")
ax[1].set_xlabel("Avg")
ax[1].set_ylabel("Run")



fig.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batter['avg'],batter['strike_rate'], color='red')
ax[0,0].set_title("Avg vs Strike rate")
ax[0,0].set_xlabel("Avg")
ax[0,0].set_ylabel("Strike rate")

ax[0,1].scatter(batter['avg'],batter['runs'], color='blue')
ax[0,1].set_title("Avg vs Run")
ax[0,1].set_xlabel("Avg")
ax[0,1].set_ylabel("Run")

ax[1,0].hist(batter['avg'])
ax[1,1].hist(batter['runs'])


fig.show()
```

```python
# one by one (alternative way)
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batter['avg'], batter['strike_rate'], color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.scatter(batter['avg'], batter['runs'], color='green')

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batter['runs'])
```
